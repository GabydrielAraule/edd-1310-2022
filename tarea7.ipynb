```python
class Nodo:
    def __init__(self, value, siguiente = None):
        self.data = value
        self.next = siguiente
class LinkedList:
    def __init__(self):
        self.head = None
    
    def is_empty(self):
        return self.head ==None

    def get_tail(self):
        if self.is_empty():
            return None
        else:
            curr_node=self.head
            while curr_node.next != None:
                curr_node = curr_node.next
            return curr_node
        
    def append(self, value):
        if self.head == None:
            self.head = Nodo(value)
        else:
            curr_node = self.head
            while curr_node.next != None:
                curr_node = curr_node.next
            curr_node.next = Nodo(value)

    def transversal(self):
        curr_node = self.head
        while curr_node != None:
            print(f"| {curr_node.data} | -->", end="")
            curr_node = curr_node.next
        print("")

    def remove(self, value):
        if self.head== None:
            print("Lista vacia no se puede remover nada")
        else:
            curr_node = self.head
            aux = None
            while curr_node.data != value:
                aux = curr_node
                curr_node = curr_node.next

            if curr_node != None:
                aux.next = curr_node.next
                curr_node = None

    def preppend(self, value):
        if self.head == None:
            print("VAcio no se puede hacer preprend")
        else: 
            aux = self.head
            self.head = Nodo( value , aux)


    def pop(self):
        if self.head == None:
            print("Lista vacia no se puede sacar ningun elemento")
        else:
            curr_node = self.head
            aux = None
            while curr_node.next != None:
                aux = curr_node
                curr_node = curr_node.next
            if curr_node != None:
                aux.next = curr_node.next
                curr_node = None

    def pop_first(self):
        if self.head == None:
            print("Lista vacia no se puede sacar ningun elemento")
        else:
            curr_node = self.head
            aux = None
            while curr_node.next != None:
                aux = self.head
                curr_node = curr_node.next         
            if curr_node != None:
                curr_node = None
                aux = curr_node
            return curr_node
     
        
```
